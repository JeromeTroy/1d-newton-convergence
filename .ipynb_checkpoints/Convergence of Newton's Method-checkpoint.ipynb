```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import diff, exp
from sympy.abc import x
from sympy.utilities.lambdify import lambdify
```

# Newton's Method

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

```python
def newton(f, f_prime, x0, f_tol=1e-16, x_tol=1e-16, maxiter=100):
    # apply newton's iteration
    
    # book keeping
    x_list = [x0]
    curr_iter = 0
    f_error = np.inf
    x_error = np.inf
    
    # newton iteration, stop when converged or gone too long
    while f_error > f_tol and x_error > x_tol and curr_iter < maxiter:
        # newton's updated
        x_curr = x_list[-1]
        x_next = x_curr - f(x_curr) / f_prime(x_curr)
        x_list.append(x_next)
        
        # update book keeping
        f_error = abs(f(x_next))
        x_error = abs(x_next - x_curr)
        curr_iter += 1
        
    return x_list
        
```

## Test 1

$$
f(x) = x^2 - 4
$$

```python
f_expr = x**2 - 4
print(f_expr)

f = lambdify(x, f_expr)
f_prime = lambdify(x, diff(f_expr, x))

x_plot = np.linspace(-3, 3, 100)
plt.figure(figsize=(13, 10))
plt.plot(x_plot, f(x_plot))
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
x0 = 5
newton_path = newton(f, f_prime, x0)
print(newton_path)

root = 2
errors = [xn - root for xn in newton_path]
for e in errors:
    print(e)
```

```python
errors_arr = np.array(errors)
print(errors_arr[1:] / errors_arr[:-1]**2)
f_double_prime = lambdify(x, diff(f_expr, [x, 2]))
print(f_double_prime(root) / (2 * f_prime(root)))
```
